```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f :
        data = json.load(f)
    return data
```

```python
def clean_data(data):
    
    #Removing users with no name
    data["users"] = [user for user in data["users"] if user["name"].strip()]
    
    #Removing Duplicate friends
    for user in data['users'] :
        user["friends"] = list(set(user['friends']))
    
    #Removing inactive users
    data['users'] = [user for user in data['users'] if user["liked_pages"] or user["friends"]]
    
    #Removing Duplicate pages
    unique_pages = {}
    for page in data['pages']:
        unique_pages[page['id']] = page
    data['pages'] = list(unique_pages.values())
    return data
```

```python
data = load_data("data2.json")
cleaned_data = clean_data(data)
json.dump(cleaned_data,open("cleaned_data2.json","w") , indent = 4)
```
