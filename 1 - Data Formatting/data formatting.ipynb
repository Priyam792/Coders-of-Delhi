```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f :
        data = json.load(f)
    return data
```

```python
def display_user(data):
    print("Users and their connections: ")
    for users in data["users"]:
        print(f"{users['name']} (ID : {users['id']}) - Friends : {users['friends']} - Liked Pages : {users['liked_pages']}")
    print("\nPages")
    for pages in data["pages"]:
        print(f"{pages['id']} : {pages['name']}")
```

```python
data = load_data("data.json")
display_user(data)
```
